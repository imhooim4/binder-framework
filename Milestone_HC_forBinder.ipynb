```python
#4. Considering only NTAs with at least 5 listed businesses, what is the average total savings and the total jobs created for each NTA?
```

```python
df_02 = df.groupby('NTA').filter(lambda x: len(x) >= 5)
```

```python
nta_summary = df_02.groupby('NTA').agg({'Total Savings': 'mean', 'Job created': 'sum'})
```

```python
print(nta_summary)
```

```python
nta_summary_02 = nta_summary.rename(columns={'Total Savings' : 'average of Total Savings','Job created' : 'sum of Job created' })
```

```python
print(nta_summary_02)
```

```python
# 1. Create a scatter plot of jobs created versus average savings. Use both a standard and a logarithmic scale for the average savings.
```

```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
#1 (1) Create the scatter plot with a standard scale for average savings
```

```python
plt.scatter(nta_summary_02['average of Total Savings'], nta_summary_02['sum of Job created'])
plt.xlabel('average of Total Savings')
plt.ylabel('sum of Job created')
plt.title('sum of Job created versus. average of Total Savings')
```

```python
#1 (2) Create the scatter plot with a logarithmic scale for average savings
```

```python
plt.scatter(nta_summary_02['average of Total Savings'], nta_summary_02['sum of Job created'])
plt.xlabel('average of Total Savings (log scale)')
plt.ylabel('sum of Job created')
plt.title('sum of Job created versus. average of Total Savings (log scale)')
plt.xscale('log')
```

```python
# 2. create a histogram of the log of the average total savings.
```

```python
# compute the log of the average total savings
log_avg_total_savings = np.log(nta_summary_02['average of Total Savings'])
```

```python
plt.hist(log_avg_total_savings)
plt.xlabel('log of average of Total Savings')
plt.ylabel('Count')
plt.title('Histogram ofaverage of Total Savings')
```

```python
# 3. Create a line plot of the total jobs created for each month
```

```python
# Convert the 'Effective Date' column to a datetime data type using pd.to_datetime().
df['Effective Date'] = pd.to_datetime(df['Effective Date'])
```

```python
# Create a 'year_month' column
df['year_month'] = df['Effective Date'].dt.strftime('%Y-%m') #strftime is "string from time"
```

```python
# Group the DataFrame by 'year_month' and sum 'Job_created' column for each 'year_month'
year_month_job_created = df.groupby('year_month')['Job created'].sum()
```

```python
year_month_job_created.describe
```

```python
# Create a line plot of the jobs created for each month
plt.plot(year_month_job_created.index, year_month_job_created.values)
plt.xlabel('Year-Month')
plt.ylabel('Total Jobs Created')
plt.title('Total Jobs Created by Year-Month')
```
